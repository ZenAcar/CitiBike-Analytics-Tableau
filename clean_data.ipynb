```python
import pandas as pd
import numpy as np
import glob
```

```python
# Concat all the 2019 Summer csv files into one dataframe

trip_2019Sum_df = pd.concat([pd.read_csv(f) for f in glob.glob('Resources/tripdata/2019/2019Sum/*.csv')], ignore_index=True)
```

```python
trip_2019Sum_df.head()
```

```python
# Check if it has duplicates
duplicate_df =trip_2019Sum_df[trip_2019Sum_df.duplicated()]
duplicate_df
```

```python
trip_2019Sum_df.count()
```

```python
# Save to csv
trip_2019Sum_df.to_csv('Resources/Cleaned_Tripdata_2019Sum.csv')
```
